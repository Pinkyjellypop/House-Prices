```python
import pandas as pd
```

```python
train_df = pd.read_csv('train.csv')
train_df
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.lmplot(x = 'LotArea', y = 'SalePrice', data = train_df, fit_reg = False, palette = "Set1")
plt.title("Lot Area vs Sale Price")
```

```python
correlation = train_df['LotArea'].corr(train_df['SalePrice'])
correlation
```

```python
freq_roof_style = train_df['RoofStyle'].value_counts()
freq_roof_style
```

```python
relative_frequencies = freq_roof_style/freq_roof_style.sum()*100
for category, frequency in relative_frequencies.items():
    print(f"{category}: {frequency:.2f}%")
```

```python
sns.barplot(x = freq_roof_style.index, y = freq_roof_style.values, palette = "Set2")
plt.title("Distribution of Roof Style")
plt.xlabel("Roof Style")
plt.ylabel("Count")
```

```python
sns.lmplot(x = 'GarageYrBlt', y = 'GarageArea', data = train_df, fit_reg = True, palette = "Set1")
plt.title("Garage Year Built vs Garage Area")
```

```python
correlation = train_df['GarageYrBlt'].corr(train_df['GarageArea'])
correlation
```
